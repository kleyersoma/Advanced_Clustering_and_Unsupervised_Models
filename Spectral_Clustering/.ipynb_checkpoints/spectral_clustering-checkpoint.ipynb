```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

from sklearn.cluster import KMeans, SpectralClustering
from sklearn.preprocessing import StandardScaler
```

```python
# Set random seed for reproducibility
np.random.seed(1000)
```

```python
NB_SAMPLES = 1000
```

```python
# Create Dataset
X = np.zeros(shape=(NB_SAMPLES, 2))
```

```python
for i in range(NB_SAMPLES):
    X[i, 0] = float(i)
    
    if i % 2 == 0:
        X[i, 1] = 1.0 + (np.random.uniform(0.65, 1.0) *
                         np.sin(float(i) / 100.0))
    else:
        X[i, 1] = 0.1 + (np.random.uniform(0.5, 0.85) *
                         np.sin(float(i) / 100.0))
```

```python
ss = StandardScaler()
Xs = ss.fit_transform(X)
```

```python
# Show the dataset
sns.set()
fig, ax = plt.subplots(figsize=(16, 8))
ax.scatter(Xs[:, 0], Xs[:, 1], marker="o", s=50)

ax.set_xlabel(r'$x_0$', fontsize=18)
ax.set_ylabel(r'$x_1$', fontsize=18)
ax.grid(True)
plt.show()
```

```python
# Test K-Means
kmeans = KMeans(n_clusters=2, random_state=1000)
Y_kmeans = kmeans.fit_predict(Xs)
```

```python
# Plot the result
sns.set()

fig, ax = plt.subplots(figsize=(16, 8))

ax.scatter(Xs[Y_kmeans == 0, 0], Xs[Y_kmeans == 0, 1], marker="o", s=100, label="Class 0")
ax.scatter(Xs[Y_kmeans == 1, 0], Xs[Y_kmeans == 1, 1], marker="d", s=100, label="Class 1")

ax.set_xlabel(r'$x_0$', fontsize=18)
ax.set_ylabel(r'$x_1$', fontsize=18)
ax.grid(True)
ax.legend(fontsize=18)
plt.show()
```

```python
# Apply Spectral clustering
sc = SpectralClustering(n_clusters=2,
                        affinity='nearest_neighbors',
                        n_neighbors=19,
                        random_state=1000)
Y_sc = sc.fit_predict(Xs)
```

```python
# Plot the result
fig, ax = plt.subplots(figsize=(16, 8))

ax.scatter(Xs[Y_sc == 0, 0], Xs[Y_sc == 0, 1], marker="o", s=100, label="Class 0")
ax.scatter(Xs[Y_sc == 1, 0], Xs[Y_sc == 1, 1], marker="s", s=100, label="Class 1")

ax.set_xlabel(r'$x_0$', fontsize=18)
ax.set_ylabel(r'$x_1$', fontsize=18)
ax.legend(fontsize=18)
ax.grid(True)
plt.show()
```
