```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from skfuzzy.cluster import cmeans, cmeans_predict
from sklearn.datasets import load_digits
```

```python
# Set random seed for reproducibility
np.random.seed(42)
```

```python
# Load the dataset
digits = load_digits()
X_train = digits['data'] / np.max(digits['data'])
```

```python
# Perform Fuzzy C-Means
fc, W, _, _, _, _, pc = cmeans(X_train.T, c=10, m=1.1, error=1e-6, maxiter=10000, seed=42)
```

```python
print('Partition coeffiecient: {}'.format(pc))
```

```python
# Plot the centroids
sns.set()
```

```python
fig, ax = plt.subplots(1, 10, figsize=(10, 10))
for i in range(10):
    c = fc[i]
    ax[i].matshow(c.reshape(8, 8) * 255.0, cmap='gray')
    ax[i].set_xticks([])
    ax[i].set_yticks([])

plt.show()
```

```python
# Membership degrees of a sample representing the digit '5'
print('Membership degrees: {}'.format(W[:, 5]))
```

```python
fig, ax = plt.subplots(figsize=(15, 8))

ax.plot(np.arange(10), W[:, 5], "o-")
ax.set_xlabel('Cluster index', fontsize=18)
ax.set_ylabel('Fuzzy membership', fontsize=18)
ax.grid(True)

plt.show()
```

```python
# Perform a prediction
new_sample = np.expand_dims(X_train[8], axis=1)
Wn, _, _, _, _, _ = cmeans_predict(new_sample, cntr_trained=fc, m=1.25, error=1e-6, maxiter=10000, seed=42)
```

```python
print('Membership degrees: {}'.format(Wn.T))
```
